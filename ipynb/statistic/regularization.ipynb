# What is $L_p$ norm?

$\forall x \in \mathbb{R}^n ~~ \forall p \in \mathbb{R} ~~s.t.~~p\geq 1$
$$ \| x\|_p = \left( \sum_{i=1}^n |x_i|^p \right)^{1/p} $$
i.e.
 $$ \| x\|_p = \left( |x_1|^p + |x_2|^p + \cdots + |x_n|^p \right)^{1/p}  $$

and if $p=\infty$
$$ \| x\|_\infty = \max\left\{ |x_1| + |x_2| + \cdots + |x_n| \right\} $$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def lp(x, p=1):
    x = np.abs(x)
    return np.sum((x**p))**(1./p)
```

```python
p = [1, 2, 4, 6, 8, 10]
plt.figure(figsize=(8, 8))
for i in range(len(p)):
    plt.subplot(2, 3, i+1)
    for j in range(500):
        x = np.array([np.random.rand()*2-1, np.random.rand()*2-1])
        if lp(x, p[i]) < 1:
            plt.scatter(x[0], x[1], s=5, c='r')
            plt.title('L_{}-norm'.format(p[i]))
            plt.axis([-1.5, 1.5, -1.5, 1.5])
            plt.tight_layout()
```
